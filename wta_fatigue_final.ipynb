## Введение

### Контекст

В последние годы элитные теннисистки часто критикуют Женскую теннисную ассоциацию (англ. Women's Tennis Association, WTA), под началом которой проводятся крупнейшие профессиональные турниры. WTA несколько раз ужесточала правила об обязательных турнирах. С <a href="https://photoresources.wtatennis.com/wta/document/2025/01/06/50d1eafa-e678-4081-bfa8-5576f670f801/2025-WTA-Rulebook-1-5-2025-.pdf">2025 года</a> теннисистки, если они попали в основную сетку, должны приезжать на все четыре турнира "Большого шлема", все десять — категории WTA-1000, шесть на выбор — WTA-500 и WTA Finals для восьмерки лучших по итогам сезона (всего 21-22 турнира). Пропуски без медицинского отвода или другой уважетельной причины наказывают денежными штрафами и потерей рейтинговых очков. 

Несмотря на санкции, некоторые игроки, например первая ракетка мира <a href="https://www.skysports.com/tennis/news/11061/13491903/aryna-sabalenka-to-skip-events-in-insane-2026-despite-likely-wta-sanctions-brits-katie-boulter-cam-norrie-suffer-defeats?utm_source=chatgpt.com">Арина Соболенко</a>, предпочитают отдохнуть (или сыграть показательные матчи с большими призовыми, которые не входят в WTA-тур). В ближайшие годы в календаре могут появиться еще один-два "тысячника"в арабских странах Персидского залива. Спортсменки опасаются, что у них добавятся новые обязательства, которые будут угрожать их здоровью и благополучию.


### Цели работы

В исследовании предпринята попытка ответить на следующие вопросы:

- Как менялась нагрузка на теннисисток и играют ли они все чаще?
- Действительно ли в женском теннисе стало больше травм и можно ли это связать с нагрузкой?
- Связана ли нагрузка с качеством игры?

### Материалы и методы

Основной источник данных — [репозиторий](https://github.com/jeffsackmann) спортивного аналитика и колумниста [Джеффа Сакманна](https://www.jeffsackmann.com/), где собраны данные по сезон 2024 года включительно. У его датасетов есть существенный недостаток: для всех матчей одного и того же турнира указана одна и та же дата. Точные даты проведения матчей доступны в датасетах с сайта <http://www.tennis-data.co.uk/>, но данные неполные. Для матчей, где дата отсутствует, она указана приблизительно с оглядкой на даты других матчей того же раунда турнира, с помощью интерполяции дат других раундов либо от даты начала турнира с учетом номера раунда. 

Нагрузка учтена двумя способами: через номер матча в сезоне (хроническая) и через продолжительность текущей игровой серии (острая). Конкретный матч отнесен к текущей серии, если он прошел в пределах четырех суток после предыдущего. Так как продолжительность матчей известна не всегда, использована прокси-метрика — количество разыгранных очков. Там, где это количество неизвестно, число геймов в матче умножено на среднее значение очков на гейм. 

Доля попаданий в корт первой подачей, доля подач навылет (эйсов), доля двойных ошибок и их отклонения от среднего по сезону для данной спортсменки взяты как перформанс-метрики, поскольку подача — первый удар в розыгрыше: она меньше зависит от действий соперницы и больше — от усталости. 

Сезонная нагрузка оценена подсчетом турниров, матчей и геймов за сезон, травмы и другие проблемы со здоровьем — подсчетом неявок на матч и отказов во время игры. Чтобы выявить связь нагрузки с перформанс-метриками и неявками/отказами, проведен регрессионный анализ. 

### Результаты

#### Нагрузка на игроков

- Несмотря на нововведения WTA, 32 теннисистки с наивысшим рейтингом по итогам сезона — те, кого правила затрагивают в первую очередь, — в целом не стали играть чаще. После сезона 2020 г., сокращенного из-за пандемии ковида, среднее количество матчей составило у них 52,6, что примерно на 7% меньше, чем в 2010-2019 гг. Реже всего лучшие спортсменки играли в первой половине 1990-х. Например, в сезоне 1995 г. типичная теннисистка (между I и III квартилями) играла от 38 до 53 матчей. Уже в 1998-м этот показатель вырос до 47-66 матчей за сезон и с тех пор колебался в узких пределах.

- Один из самых громких критиков WTA Ига Швентек в трех последних сезонах играла чаще всех остальных: 76, 82, 74 матча в одиночных турнирах. 
- Средняя продолжительность матчей, оцененная в сыгранных геймах, в течение всего рассмотренного периода практически не менялась. 
- Количество длинных игровых серий (восемь и более матчей с короткими перерывами: семь нужно, чтобы выиграть турнир "Большого шлема") рассчитано с 2007 г. До пандемии ковида наблюдался тренд на снижение, а после нее — и нововведений WTA — количество таких серий снова увеличилось. Тем не менее за последние три сезона их ни разу не набралось хотя 30 на всех — это меньше, чем в любом сезоне до 2015 г. включительно.
- Доля матчей в длинных сериях от общего количества матчей в сезоне теперь также далека от исторических максимумов. В 2007-2013 гг. этот показатель превышал 20% пять раз из семи, в 2020-х — ни разу, а в 2024 г. составил 13,1%.
- В целом, расчеты не подтверждают, что в последнее время нагрузка на теннисисток увеличилась. Впрочем, даже если количество матчей меняется несущественно, их интенсивность с годами может расти. Если мяч летит быстрее, а рывки надо делать чаще, то усталость будет накапливаться быстрее. Имеющиеся данные не позволяют проверить этот эффект. Кроме того, и без дополнительного увеличения нынешняя нагрузка может быть чрезмерной.

#### Неявки на матч и снятия из-за травм

- На протяжении 40 сезонов с 1984 по 2024 гг. росла доля матчей, которые заканчиваются из-за отказа спортсменки продолжать борьбу. В последние десять лет их почти всего было более 3% (втрое больше, чем в 1980-х). Как правило, причиной отказа служат травмы или плохое самочувствие, что может быть связано с физиологической и психической нагрузкой. 
- Наоборот, доля матчей, не состоявшихся из-за неявки (зачастую по тем же причинам), долго оставалась довольно стабильной. Лишь в 2020-х наметился возрастающий тренд (0,88% против 0,32-0,52% в предшествующие декады). Можно допустить, что рост первого показателя отчасти происходит за счет второго: не исключено, что теннисистки стали чаще выходить на корт не в лучшем состоянии, тогда как раньше они просто не явились бы на матч. Однако имеющиеся данные не позволяют проверить эту гипотезу.
- В 2020 году произошел естественный эксперимент: из-за пандемии ковида снизились игровая нагрузка (соревнования приостановили примерно на пять месяцев) и количество перелетов (некоторые турниры затем проводили в одном городе друг за другом). В том сезоне кратно сократились и количество снятий, и доля таких случаев, а доля неявок, наоборот, увеличилась (возможная дополнительная причина — положительный тест на SARS-CoV-2 перед матчем).


- Регрессионный анализ матчей в 2020 г., четырех предшествующих и четырех последующих сезонов (более ранние сезоны исключены из анализа, чтобы снизить влияние других факторов) показал, что в начале пандемии риск сняться во время матча был действительно ниже (OR: 0,46, CI: 0,29-0,74, p = 0,001). Из-за малой выборки статистически значимое влияние на риск неявки не обнаружено (OR: 0,91, CI: 0,44-1,86, p = 0,791). 
- На риски также влияет усталость, накопленная за предшествующие матчи без перерыва хотя бы в четыре дня. Каждый матч-эквивалент (140 разыгранных очков) повышает и риск снятья (OR: 1,12, CI: 1,08-1,17, p < 0,001), и риск не явиться (OR: 1,33, CI: 1,27-1,40, p < 0,001). Впрочем, на практике абсолютные риски все равно остаются низкими. 
- При анализе матчей только 32 теннисисток с самым высоким рейтингом по итогам сезона выявлены схожие закономерности, но величины эффектов меньше, а статистически значим только один результат: по мере накопления усталости игроки чаще не выходили на корт (OR: 1,18, CI: 1,10-1,27, p < 0,001).

- Для интерпретации важны два существенных ограничения. Первое — в неявках учтены только случаи, произошедшие на соревнованиях. Но бывает, что спортсменки снимаются еще до их начала. Например, в 2024 году на одних только турнирах "Большого шлема" (всего их четыре за сезон) это случилось девять раз. Теннисистку, снявшуюся перед турниром, заменяет та, кто имеет наивысший рейтинг среди проигравших в финалах квалификации. В сетке ее помечают как "везучую неудачницу" (англ. lucky loser, общеупотребимое сокращение — LL). Зная количество LL, можно было бы точнее оценить динамику неявок по сезонам, но в имеющихся данных в этом поле почти везде пропуски.
- Второе существенное ограничение — данные не позволяют судить о тяжести травм и о том, новое это повреждение или рецидив хронической проблемы. И то, и другое способно повлиять как на динамику показателей по сезонам, так и на их интерпретациию.  

#### Перформанс-метрики

- Регрессионный анализ дал статистически значимые результаты для нескольких факторов, влияющих на метрики подачи, но сила эффектов невелика. Например, каждый матч-эквивалент накопленной нагрузки в серии связан с изменением доли попавших в корт первых подач всего на 0.3 п.п. Интересно, что, согласно модели, по мере накопления нагрузки точность подач увеличивается, а двойные ошибки случаются реже. Это можно объяснить тем, что более поздние матчи в серии обычно приходятся на последние стадии турниров, где требуется повышенная концентрация, а цена ошибки выше.

- Сила соперницы на приеме и брейк-поинты, когда каждая ошибка стоит очка, на практике также не связаны с метриками подачи, хотя некоторые полученные результаты имеют статистическую значимость. 
- Имеющиеся данные не подтверждают, что нагрузки в современном женском теннисе связаны с ухудшением выступлений спортсменок. 
- Дополнительная информация позволила бы усовершенствовать модели. Так, накопленная нагрузка оценена через прокси-метрику - предпочтительнее было бы измерять ее в часах на корте, а также учитывать время между матчами, перелеты и другие факторы усталости.

Хотя не все результаты согласуются с доводами критиков WTA, дальнейшие исследования в этой области выглядят оправданными, особенно в том, что касается травм и временной потери трудоспособности. Лучше понять, как усталость сказывается на теннисистках, позволили бы данные о ходе матчей (а не только итоговая статистика), показатели фитнес-трекеров и биомаркеры спортсменок. Не приходится надеяться, что сведения о пульсе или метаболическом профиле когда-нибудь будут доступны для стороннего анализа, однако в интересах WTA провести такие исследования.

## Функции

```python
# Функция для подсчета пропусков.
def missing_value_summary(df):

    # Считаем.
    total_rows = len(df) # Количество строк в датафрейме.
    missing_counts = df.isnull().sum() # Количество пропусков в каждом столбце.
    missing_proportion = (missing_counts / total_rows) * 100 # Доля пропусков в процентах.
    unique_counts = df.nunique() # Количество уникальных значений в каждом столбце.
    
    # Сводим.
    summary_df = pd.DataFrame({
        '№': range(1, len(df.columns) + 1),
        'Тип данных': df.dtypes,
        'Уникальные': unique_counts,
        'Пропуски_число': missing_counts, 
        'Доля, %': missing_proportion
    })

    # Дополняем.
    summary_df['Пропуски'] = summary_df['Пропуски_число'].astype(str) + ' / ' + str(total_rows) # Выводим дробью.
    summary_df['Excel'] = [get_column_letter(i) for i in range(1, len(df.columns) + 1)] # Указываем столбцы в формате Excel.
    
    # Приводим в порядок.
    cols_order = ['№', 'Excel', 'Тип данных', 'Уникальные', 'Пропуски', 'Доля, %']
    summary_df = summary_df[cols_order]

    # Делаем красиво.
    summary = (
        summary_df.style
        .format({'Доля, %': '{:.2f}'})
        .background_gradient(cmap='Reds', subset=['Доля, %'], vmin=0, vmax=100)
        .map(lambda val: 'background: limegreen; color: white' if val == 0 else '', subset=['Доля, %'])
        .set_table_attributes('style="display: block; max-height: 450px; overflow-y: auto; border: 1px solid #ccc;"')
    )

    return summary
```

```python
# Функция для очистки счета, подсчета количества сетов и геймов в матче.
def parse_score(score_str):

    # Приводим к строке и в верхний регистр.
    s = str(score_str).upper()

    # Удаляем артефакты парсинга (?) вроде +H61 или H61.
    s = re.sub(r'\+?H\d+', '', s)
    
    # Если одна из теннисисток не вышла на матч, то столбцы с количеством сетов и геймов заполняем нулями.
    if any(x in s for x in ['W/O', 'WALKOVER']):
        return s, 0, 0
    
    ### Очищаем счет.
    # Чистим сеты с тай-брейками: 7-6(5) ---> 7-6.
    clean_s = re.sub(r'\([^)]*\)', '', s)

    # Обрабатываем чемпионский тай-брейк: [10-6] ---> 1-0.
    clean_s = re.sub(r'\[.*?\]', '1-0', clean_s)

    # Чиним сшитые сеты типа 6-0-6-1.
    clean_s = re.sub(r'(\d+-\d+)-(\d+)', r'\1 \2', clean_s)

    # Оставляем только цифры, дефисы и пробелы
    clean_s = re.sub(r'[^0-9\-\s]', '', clean_s)

    # Убираем лишние пробелы: 6- 0 ---> 6-0
    clean_s = clean_s.replace('- ', '-').replace(' -', '-').strip()
    clean_s = re.sub(r'\s+', ' ', clean_s)
    
    # Разбиваем на сеты по пробелам.
    sets_data = clean_s.split()
    
    # Считаем сеты.
    sets_count = len(sets_data)
    
    # Считаем геймы.
    games_count = sum(int(n) for n in re.findall(r'\d+', clean_s))

    return clean_s, sets_count, games_count
```

```python
# Функция для генерации ключа из имени теннисистки. Ключ требуется для присоединения к датафрейму с данными Сакманна.
def generate_name_key_td(series):
    
    s = series.str.strip() # Теннисистки указаны как Фамилия И. или Фамилия И.И. 
    before_dot = s.str.split('.').str[0].str.strip() # Берем все до первой точки.    
    initial = before_dot.str[-1].str.lower() # Выделяем инициал, переводим в нижний регистр.
    last_name = before_dot.str[:-1].str.strip().str.lower().str.replace(' ', '_') # Выделяем фамилию.
    
    return last_name + "_" + initial # Собираем ключ.
```

```python
# Еще один генератор ключа из имени и фамилии теннисистки, но уже в данных Сакманна.
def generate_name_key_sackmann(last_name, first_name):

    last = str(last_name).lower().strip().replace(' ', '_') # Фамилии бывают двойные.
    initial = str(first_name).strip()[0].lower() if pd.notna(first_name) and str(first_name).strip() else ""
    return f"{last}_{initial}"
```

```python
# Функция для трансформации таблицы. Данные разворачиваются с опорой на флаг победильницы, который всегда есть у одной теннисистки и отсутствует у другой. 
def extract_player_data(is_winner):

    # Общие колонки.
    shared_cols = [
        'tourney_id', 'tourney_level', 'surface', 'match_id', 'date', 'year', # Турнир и матч.
        #'is_2016_onwards', 
        'is_ret', 'is_wo', 'score', # Флаги и счет.
        'sets_played', 'games_played', 'points_played' # Продолжительность.
    ]

    # Игровые показатели теннисисток.
    stats_list = ['ace', 'df', 'svpt', '1stIn', '1stWon', '2ndWon', 'SvGms', 'bpSaved', 'bpFaced']
    
    p_long, o_long = ('winner', 'loser') if is_winner else ('loser', 'winner')
    p_short, o_short = ('w', 'l') if is_winner else ('l', 'w')
    
    mapping = {
        f"{p_long}_id": "player_id",
        f"{o_long}_id": "opponent_id",
        f"{p_long}_name": "player_name",
        f"{o_long}_name": "opponent_name",
        f"{p_long}_age": "age",
        f"{p_long}_rank": "rank",
        f"{o_long}_rank": "opp_rank"
    }
    
    # Собственные игровые показатели.
    for s in stats_list:
        mapping[f"{p_short}_{s}"] = s
        
    # Показатели соперницы.
    for s in stats_list:
        mapping[f"{o_short}_{s}"] = f"opp_{s}"
        
    # Предохранитель.
    valid_cols = [c for c in shared_cols if c in df_matches.columns]
    valid_mapping = {k: v for k, v in mapping.items() if k in df_matches.columns}
    
    df = df_matches[valid_cols + list(valid_mapping.keys())].rename(columns=valid_mapping)
    df['won'] = is_winner
    
    return df
```

Склейка по трем ключам оставляет маленькую, но не нулевую вероятность ошибочного совпадения. Ошибка возникнет в тех случаях, когда одна теннисистка обыграла другую несколько раз с одинаковым счетом. Например, Агнешка Радваньская победила Аи Сугияму со счетом 6-2 6-1 сначала 5 августа 2009 г. в Лос-Анджелесе, а потом 11 августа 2009 г. в Цинцинатти. Совпадения вроде этого чрезвычайно редки. Тем примечательнее, что с Радваньской такое случалось дважды.

Финальная проверка в ячейке ниже нужна вот зачем. Представим, что для первого матча между Радваньской и Сигуямой даты нет, а для второго есть. Тогда к первому присоединится только дата второго, и дубль не возникнет. Следовательно, ошибочная строка не отфильтруется, а дата останется неверная. Код проверяет, что дата матча находится в разумных пределах от даты начала турнира у Сакманна, и затирает неверные даты. 

По иронии, именно в случае с Радваньской и Сугиямой проверка не помогла бы, так как исход и счет в их матчах не просто совпали, но еще и с разницей всего в шесть дней. Если бы во вспомогательном датафрейме для этой пары матчей не хватало бы одной и только одной даты, то это была бы трижды редкость. Чтобы учесть этот случай, можно было бы сузить окно соответствия в маске. Но, во-первых, немало турниров длятся 12-14 дней, во-вторых, если бы Радваньска и Сугияма сначала встретились на поздней стадии турнира, а потом на ранней, то не помогла бы и эта мера. Остается только смириться с риском. Кроме того, далее в коде проводится проверка хронологического порядка дат. 

```python
# Считаем разницу датой, указанной во вспомогательном датафрейме, и датой у Сакманна.
df_matches['date_diff'] = (df_matches['date'] - df_matches['tourney_date']).dt.days

# Считаем разницу по модулю.
df_matches['abs_diff'] = df_matches['date_diff'].abs()

# Сортируем так, чтобы сначала были строки с наименьшей разницей (то есть те, где склейка с большей вероятностью правильная).
df_matches = df_matches.sort_values(['tourney_id', 'winner_id', 'loser_id', 'round', 'abs_diff']) # Раунд нужен для итоговых турниров, где игроки могут сначала встретиться в группе, а потом в плей-офф.

# Удаляем дубликаты, появившиеся из-за неверного присоединения.
df_matches = df_matches.drop_duplicates(subset=['tourney_id', 'winner_id', 'loser_id', 'round'], keep='first')

# Последняя проверка. 
mask = (df_matches['date_diff'] >= -1) & (df_matches['date_diff'] <= 14)
df_matches.loc[~mask, 'date'] = pd.NaT

df_matches = df_matches.drop(columns=['date_diff', 'abs_diff'])
```

Предшествующие операции позволили получить даты проведения лишь для части матчей. Оставшиеся пропуски можно заполнить с приемлемой точностью, опираясь на имеющиеся даты и порядок раундов в турнирах. 
- Если известно, когда прошли две трети матчей первого круга, то справедливо допустить, что еще треть состоялась тогда же; пропуски заполняются пропорционально распределению известных дат. 
- Если в раунде совсем нет дат, то справедливо допустить, что они лежат где-то между датами предшествующего и последующих раундов; пропуски заполняются с помощью интерполяции. 
- Если нет никаких дат, кроме указанной Сакманном, то шагаем от нее с учетом порядка раундов.

```python
# Указываем раунды числами в обратном порядке.
round_order = {'F': 0, 'SF': 1, 'QF': 2, 'R16': 3, 'R32': 4, 'R64': 5, 'R128': 6, 'RR': 2, 'BR': 0}
df_matches['round_num'] = df_matches['round'].map(round_order)

# Вычисляем среднюю дату раундов в каждом турнире.
round_dates = df_matches.groupby(['tourney_id', 'round_num'])['date'].mean().reset_index()
round_dates = round_dates.sort_values(['tourney_id', 'round_num'])

# Проверяем, нет ли нарушений порядка. 
round_dates['prev_date'] = round_dates.groupby('tourney_id')['date'].shift(-1)
violations = round_dates[round_dates['date'] < round_dates['prev_date']]

if not violations.empty:
    print(f"Найдено хронологических нарушений: {len(violations)}")
    print(violations[['tourney_id', 'round_num', 'date', 'prev_date']].head())

# Интерполируем даты для раундов, где вообще нет данных.
round_dates['date'] = round_dates.groupby('tourney_id')['date'].transform(
    lambda x: x.interpolate(method='linear', limit_direction='both')
)

# Мапим. 
df_matches = df_matches.merge(
    round_dates.rename(columns={'date': 'interp_date'}).drop(columns=['prev_date']), 
    on=['tourney_id', 'round_num'], 
    how='left'
)

# Функция для последовательного заполнения дат. 
def fill_dates(group):
    if group['date'].notna().any():
        counts = group['date'].value_counts()
        n_missing = group['date'].isna().sum()
        if n_missing > 0:
            fill_values = np.random.choice(counts.index, size=n_missing, p=(counts/counts.sum()).values)
            group.loc[group['date'].isna(), 'date'] = fill_values
    
    group['date'] = group['date'].fillna(group['interp_date'])
    
    max_round_in_tourney = group['round_num'].max()
    offset = (max_round_in_tourney - group['round_num']).astype('timedelta64[D]')
    group['date'] = group['date'].fillna(group['tourney_date'] + offset)
    
    return group

# Применяем.
df_matches = df_matches.groupby(['tourney_id', 'round'], group_keys=False).apply(fill_dates)

df_matches = df_matches.sort_values(['tourney_id', 'date', 'round_num'])
df_matches = df_matches.drop(columns=['interp_date', 'round_num'])
```

```python
# Для пропусков в рейтингах на момент проведения матча пригодится вспомогательная таблица Сакманна.

# Подготовка.
df_main['date'] = pd.to_datetime(df_main['date'])
df_wta_rankings['ranking_date'] = pd.to_datetime(df_wta_rankings['ranking_date'], format='%Y%m%d')

df_main = df_main.sort_values('date')
df_wta_rankings = df_wta_rankings.sort_values('ranking_date')

# Присоединяем рейтинг из внешней таблицы.
df_main = pd.merge_asof( # Так как рейтинг обновляется раз в неделю, а матчи проходят чаще. 
    df_main, 
    df_wta_rankings[['ranking_date', 'player_id', 'rank']].rename(columns={'rank': 'ext_rank'}),
    left_on='date', 
    right_on='ranking_date', 
    by='player_id', 
    direction='backward'
)

# Сначала ищем рейтинг в данных из вспомогательной таблицы.
df_main['rank'] = df_main['rank'].fillna(df_main['ext_rank'])

# Если не нашлось, то заполняем ближайшим известным. 
df_main['rank'] = df_main.groupby('player_id')['rank'].ffill().bfill()

# Если вообще ничего нет, ставим значение по умолчанию. 
df_main['rank'] = df_main['rank'].fillna(1000)

# Так как теперь на матч приходится по две строки, заполняем зеркально.  
df_main['opp_rank'] = df_main.groupby('match_id')['rank'].transform(lambda x: x.iloc[::-1].values)

df_main = df_main.drop(columns=['ext_rank', 'ranking_date'])
```

#### Разметка игровых серий

Игровая серия — это последовательность матчей, которые идут друг за другом с короткими перерывами. Например, матчи на одном турнире — это серия. Но если теннисистка выиграла турнир в выходные, а в понедельник вынуждена играть на следующем, то это считается продолжением серии. Предположительно, чем длиннее серия, тем больше накапливается острая усталость.

```python
df_main = df_main.sort_values(['player_id', 'date'])

# Считаем разрыв между матчами одного игрока.
df_main['diff'] = df_main.groupby('player_id')['date'].diff().dt.days

# Серия продолжается, если следующий матч сыгран в пределах четырех суток после предыдущего. 
df_main['is_new_streak'] = (df_main['diff'] > 4) | (df_main['diff'].isna())
df_main['streak_id'] = df_main.groupby('player_id')['is_new_streak'].cumsum()

# Нумеруем матчи внутри серии и рассчитываем ее длину. 
df_main['match_in_streak'] = df_main.groupby(['player_id', 'streak_id']).cumcount() + 1
df_main['streak_length'] = df_main.groupby(['player_id', 'streak_id'])['match_id'].transform('count')

# Рассчитываем накопленную нагрузку за серию в геймах.
df_main['streak_pts_before'] = (
    df_main['points_played']
    .fillna(df_main['games_played'] * 6.55)
    .groupby([df_main['player_id'], df_main['streak_id']])
    .transform(lambda x: x.cumsum().shift(1, fill_value=0))
)

df_main['opp_streak_pts_before'] = df_main.groupby('match_id')['streak_pts_before'].transform(lambda x: x.iloc[::-1].values)
```

```python
# Для каждой теннисистки нумеруем матчи по сезонам. 
df_main = df_main.sort_values(['player_id', 'date', 'match_id'])
df_main['match_num_season'] = df_main.groupby(['player_id', 'year']).cumcount() + 1
```

#### Расчет перформанс-метрик

Подача — первое действие в розыгрыше, поэтому она напрямую не зависит от действий соперницы. Тем не менее подающая может учитывать особенности соперницы при игре на приеме, отчего подача станет более или менее рискованной, а это, в свою очередь, может повлиять на статистику в конкретном матче. 

```python
# Подача: попадание первой подачей, эйсы и двойные ошибки в пересчете на количество подач.
df_main['1stIn_share'] = np.where(
    df_main['svpt'] > 0,
    df_main['1stIn'] / df_main['svpt'],
    np.nan
)

df_main['ace_share'] = np.where(
    (df_main['svpt'] - df_main['df']) > 0,
    df_main['ace'] / (df_main['svpt'] - df_main['df']),
    np.nan
)

df_main['df_share'] = np.where(
    (df_main['svpt'] - df_main['1stIn']) > 0,
    df_main['df'] / (df_main['svpt'] - df_main['1stIn']),
    np.nan
)
```

```python
# Рассчитываем усредненные показатели игроков за сезон на подаче.
season_serve_stats = (
    df_main.groupby(['player_id', 'year'], as_index=False)
    .agg({
        'svpt': 'sum',
        '1stIn': 'sum',
        '1stWon': 'sum',
        '2ndWon': 'sum',
        'ace': 'sum',
        'df': 'sum'
    })
)

season_serve_stats = season_serve_stats.rename(columns={
    col: f"{col}_season" for col in season_serve_stats.columns if col not in ['player_id', 'year']
})

season_serve_stats['svpt_won_season_share'] = np.where(
    season_serve_stats['svpt_season'] > 0,
    (season_serve_stats['1stWon_season'] + season_serve_stats['2ndWon_season']) / season_serve_stats['svpt_season'],
    np.nan
)

season_serve_stats['1stIn_season_share'] = season_serve_stats['1stIn_season'] / season_serve_stats['svpt_season']
season_serve_stats['ace_season_share'] = np.where(
    season_serve_stats['svpt_season'] - season_serve_stats['df_season'] > 0,
    season_serve_stats['ace_season'] / (season_serve_stats['svpt_season'] - season_serve_stats['df_season']),
    np.nan
)

season_serve_stats['df_season_share'] = np.where(
    season_serve_stats['svpt_season'] - season_serve_stats['1stIn_season'] > 0,
    season_serve_stats['df_season'] / (season_serve_stats['svpt_season'] - season_serve_stats['1stIn_season']),
    np.nan
)
```

```python
# Рассчитываем усредненные показатели игроков за сезон на приеме. 
# Игра соперницы на приеме может влиять на то, как игрок будет подавать. 
season_return_stats = df_main.groupby(['opponent_id', 'year']).agg({
    'svpt': 'sum',
    '1stWon': 'sum',
    '2ndWon': 'sum'
}).reset_index()

season_return_stats = season_return_stats.rename(columns={
    'svpt': 'rtpt_season',
    '1stWon': '1st_rtn_lost_season',
    '2ndWon': '2nd_rtn_lost_season'
})

season_return_stats['rtn_lost_season_total'] = season_return_stats['1st_rtn_lost_season'] + season_return_stats['2nd_rtn_lost_season']
season_return_stats['rtn_lost_season_share'] = np.where(
    season_return_stats['rtpt_season'] > 0,
    season_return_stats['rtn_lost_season_total'] / season_return_stats['rtpt_season'],
    np.nan
)
```

#### Обсуждение результатов

- Существенное ограничение — в неявках учтены только случаи, произошедшие на соревнованиях. Но бывает, что спортсменки снимаются еще до их начала. В сетке ее помечают как "везучую неудачницу" (англ. lucky loser, общеупотребимое сокращение — LL). Зная количество LL, можно было бы точнее оценить динамику неявок по сезонам, но в имеющихся данных в этом поле почти везде пропуски.

В последние десять лет доля таких матчей за сезон почти всегда превышает 3% (втрое больше, чем в 1980-х). Как правило, причиной отказа служат травмы или плохое самочувствие, что может быть связано с физиологической и психической нагрузкой. 
- Наоборот, доля матчей, не состоявшихся из-за неявки (зачастую по тем же причинам), долго оставалась довольно стабильной. Лишь в 2020-х наметился возрастающий тренд (0,88% против 0,32-0,52% в предшествующие декады). Можно допустить, что рост первого показателя отчасти происходит за счет второго: не исключено, что теннисистки стали чаще выходить на корт не в лучшем состоянии, тогда как раньше они просто не явились бы на матч. Однако имеющиеся данные не позволяют проверить эту гипотезу.
- В 2020 году произошел естественный эксперимент: из-за пандемии ковида снизились игровая нагрузка (соревнования приостановили примерно на пять месяцев) и количество перелетов (некоторые турниры затем проводили в одном городе друг за другом). В том сезоне кратно сократились и количество снятий, и доля таких случаев, а доля неявок, наоборот, увеличилась (возможная дополнительная причина — положительный тест на SARS-CoV-2 перед матчем).
- Регрессионный анализ матчей в 2020 г., четырех предшествующих и четырех последующих сезонов (более ранние сезоны исключены из анализа, чтобы снизить влияние других факторов) показал, что в начале пандемии риск сняться во время матча был действительно ниже (OR: 0,46, CI: 0,29-0,74, p = 0,001). Впрочем, на практике абсолютные риски все равно остаются низкими.

### Игровые корреляты усталости

```python
# Делаем срез для регрессионного анализа.
df_ftg = df_main[
    (df_main['year'].between(2016, 2024))
    #& (df_main['is_top32'] == 1)
    & (df_main['1stIn_share'].notna())
    & (df_main['ace_share'].notna())
    & (df_main['df_share'].notna())
    ]
```

```python
# Рассчитываем коэффициенты корреляции. 
df_ftg[[
    '1stIn_dev', 
    'ace_dev', 
    'df_dev', 
    'opp_rtn_lost_season_share', 
    'bpFaced',
    'streak_pts_before_norm', 
    'match_num_season'
    ]].corr() 
```

```python
# Зависимые переменные — метрики подачи: отклонения от среднего по сезону для попадания первой подачей, эйсов и двойных ошибок.
y_vars = ['1stIn_dev', 'ace_dev', 'df_dev']

# Независимые переменные — накопленная нагрузка в серии, за сезон, количество брейк-поинтов на своей подаче, доля очков в среднем за сезон, выигранных соперницей на приеме.
x_vars = ['streak_pts_before_norm', 'match_num_season', 'bpFaced', 'opp_rtn_lost_season_share']

# Указываем величины для конвертации коэффициентов в интутивно более понятные единицы. 
scales = {
    'streak_pts_before_norm': 1,       # Один матч-эквивалент.
    'match_num_season': 1,             # Каждый следующий матч в сезоне.
    'bpFaced': 1,                       # Один брейк-пойнт.
    'opp_rtn_lost_season_share': 0.01   # Дополнительный 1% очков, выигранных соперницей на приеме. 
}

results = []

for metric in y_vars:
    X = df_ftg[x_vars]
    X = sm.add_constant(X)
    y = df_ftg[metric]

    # Обычная модель.
    model = sm.OLS(y, X).fit()
    
    # Модель с кластеризацией по игроку.
    model_cluster = sm.OLS(y, X).fit(cov_type='cluster', cov_kwds={'groups': df_ftg['player_id']})
    
    for var in x_vars:
        coef = model.params[var]
        pval = model.pvalues[var]
        pval_cluster = model_cluster.pvalues[var]
        
        # Эффект в понятных единицах.
        effect = coef * scales[var] * 100 
        
        results.append({
            'Метрика': metric,
            'Фактор': var,
            'Эффект, п.п.': round(effect, 3),
            'p-value': round(pval, 3),
            'p-value (кластер)': round(pval_cluster, 3)
        })

df_results = pd.DataFrame(results)
df_results
```

#### Обсуждение результатов

- Регрессионный анализ дал статистически значимые результаты для нескольких факторов, влияющих на метрики подачи, но сила эффектов невелика. Например, каждый матч-эквивалент накопленной нагрузки в серии связан с изменением доли попавших в корт первых подач всего на 0.3 п.п. Интересно, что, согласно модели, по мере накопления нагрузки точность подач увеличивается, а двойные ошибки случаются реже. Это можно объяснить тем, что более поздние матчи в серии приходятся на последние стадии турниров, где требуется повышенная концентрация, а цена ошибки выше.
- Сила соперницы на приеме и брейк-поинты, когда каждая ошибка стоит очка, на практике также не связаны с метриками подачи, хотя некоторые полученные результаты имеют статистическую значимость. 
- Имеющиеся данные не подтверждают, что нагрузки в современном женском теннисе связаны с ухудшением выступлений спортсменок. 
- Дополнительная информация позволила бы усовершенствовать модели. Так, накопленная нагрузка оценена через прокси-метрику - лучше было бы измерять ее в часах на корте, а также учитывать время между матчами, перелеты и т.п. 

## Заключение и рекомендации

- Самый существенный рост количества матчей у 32 лучших спортсменок по итогам сезона случился во второй половине 1990-х годов, то есть задолго до введения правил об обязательных турнирах. В 2020-х годах нагрузка на элитных игроков была даже ниже на 7%, чем в 2010-х. Количество длинных серий и их доля за сезон в последние годы увеличиваются, но не достигают исторических максимумов. 
- Все чаще матчи заканчиваются из-за того, что одна из теннисисток больше не может играть. В 2020 г., когда играли намного меньше обычного, доля снятий резко уменьшилась, а по мере накопления нагрузки в рамках серии риск увеличивается. 
- Связь между нагрузкой и успешными подачами как индикатором состояния теннисистски слабая: статистически значимые эффекты выявлены, но на практике ими можно пренебречь. 
- Исследование имеет несколько существенных ограничений: нагрузка сама по себе не говорит об усталости, а вдобавок складывается не только из того, что происходит на корте; неявка или снятие зачастую обусловлены проблемами со здоровьем, но об их характере и тяжести судить невозможно; с годами меняется сама игра. Хотя не все результаты согласуются с доводами критиков WTA, дальнейшие исследования в этой области выглядят оправданными, особенно в том, что касается травм и временной потери трудоспособности. 
- Лучше понять, как усталость сказывается на теннисистках, позволили бы данные о ходе матчей (а не только итоговая статистика), показатели фитнес-трекеров и биомаркеры спортсменок. Не приходится надеяться, что сведения о пульсе или метаболическом профиле когда-нибудь будут доступны для стороннего анализа, однако в интересах WTA провести такие исследования.
